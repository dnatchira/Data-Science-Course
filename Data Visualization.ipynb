```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 2, 1) #(row, columns, position)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.subplot(2, 2,3)
plt.plot(x2, y2, 'b.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.subplot(2, 2, 4)
plt.plot(x1, y1, 'g+-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
plt.rcParams["figure.figsize"] = (10, 5)
plt.show()
```

```python
# Assignment
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

score = pd.read_csv("Raw_Score.csv")
grade = score["Grade"].value_counts()

plt.subplot(2,2,1) #ดึงข้อมูลจากไฟล์ Raw_Score.csv
Grade = ('A', 'B+', 'B', 'C+', 'C', 'D+', 'D', 'F')
Student = (grade['A'], grade['B+'], grade['B'], grade['C+'], grade['C'], grade['D+'], grade['D'], grade['F'])
index = np.arange(len(Grade))
plt.bar(index, Student, color="orange", align='center', alpha=0.85)
plt.xticks(index, Grade)
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.title('Grade Distribution')

Range = score["Range"].value_counts()

plt.subplot(2,2,2) #ดึงข้อมูลจากไฟล์ Raw_Score.csv โดยสร้างคอลัมน์ Range เพิ่มในไฟล์
ranges = ('0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100')
number = (Range['0-10'], Range['11-20'], Range['21-30'], Range['31-40'], Range['41-50'], Range['51-60'], Range['61-70'], Range['71-80'], Range['81-90'], Range['91-100'])
column = np.arange(len(ranges))
plt.bar(column, number, color="lightblue", align='center', alpha=0.85)
plt.xticks(column, ranges)
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Total Score Range Distribution')

plt.scatter(ranges, number,color="b", edgecolors="k", s = 30, alpha = 0.7)
plt.plot(ranges, number)

plt.subplot(2,2,3) #ป้อนค่าเฉลี่ยคะแนนรวมของแต่ละกลุ่ม จากการคำนวณ
Test = ('A', 'B+', 'B', 'C+', 'C', 'D+', 'D', 'F')
Mean = [85.02, 76.76, 71.92, 66.88, 61.93, 57.02, 51.99, 35.73]
plt.plot(Test, Mean, color='yellowgreen')
plt.scatter(Test, Mean, color="green", edgecolors="k", s = 30, alpha = 0.7)
plt.xlabel('Grade Group')
plt.ylabel('Total Score Average')
plt.title('Total Score Average of Each Grade Group')

plt.subplot(2,2,4) #ป้อนค่าเฉลี่ยคะแนนแต่ละส่วน ของแต่ละกลุ่ม จากการคำนวณ
n_groups = 8
Assignment = (11.17, 15.52, 16.71, 18.28, 19.42, 20.52, 22.22, 25.02)
Midterm = (12.23, 18.23, 20.08, 21.60, 23.63, 25.64, 27.26, 30.02)
Final = (12.32, 18.22, 20.21, 22.05, 23.81, 25.73, 27.28, 29.97)
Assignment_and_Midterm = (23.40, 33.75, 36.79, 39.88, 43.05, 46.16, 49.48, 55.04)

group = np.arange(n_groups)

plt.barh(group, Assignment, align='center', alpha=0.85,
color='lightpink',
label='Assignment')

plt.barh(group, Midterm, left=Assignment, align='center', alpha=0.85,
color='orchid',
label='Midterm')

plt.barh(group, Final,left= Assignment_and_Midterm, align='center', alpha=0.85,
color='darkmagenta',
label='Final')

plt.ylabel('Grade Group')
plt.xlabel('Score')
plt.title('Mean Score Separation of Each Grade Group')
plt.yticks(group, ('F', 'D', 'D+', 'C', 'C+', 'B', 'B+', 'A'))
plt.legend()

plt.tight_layout()
plt.rcParams["figure.figsize"] = (14, 8)
plt.show()
```
